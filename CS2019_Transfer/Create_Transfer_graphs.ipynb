```python
import sys
sys.path.append('../')
import matplotlib.pyplot as plt
import plotly
import numpy as np
import random
import hrr2 as hrr
import math
import state_machine_class as state_machine
import Gate_Class_v4 as gate
from plotly.graph_objs import Scatter, Layout, Surface, Bar, Figure
plotly.offline.init_notebook_mode(connected=True)
```

```python
def create_transfer_graph():
    
    fname0 = "I_model_TransferTask_error.dat" # Input model transfer task error data file name
    fname1 = "IO_model_TransferTask_error.dat" # Output model transfer task error data file name
    fname2 = "IO_model_baselineTask_error.dat" # Output baseline task error data file name
    '''
    fname0 = "I_model_TransferTask_error_s256.dat" # Input model transfer task error data file name
    fname1 = "IO_model_TransferTask_error_s256.dat" # Output model transfer task error data file name
    fname2 = "IO_model_baselineTask_error_s256.dat" # Output baseline task error data file name
    
    fname0 = "I_model_TransferTask_accuracy_s1024.dat" # Input model transfer task error data file name
    fname1 = "IO_model_TransferTask_accuracy_s1024.dat" # Output model transfer task error data file name
    fname2 = "IO_model_baselineTask_accuracy_s1024.dat" # Output baseline task error data file name
    '''
    f_obj0 = open(fname0,"r")
    f_obj1 = open(fname1,"r")
    f_obj2 = open(fname2,"r")
    
    I_model_transfer = np.array(f_obj0.readlines())
    IO_model_transfer = np.array(f_obj1.readlines())
    IO_model_baseline = np.array(f_obj2.readlines())
    
    size = 25
    ticktext = [x for x in range(0,50000,200)]
    tickvals = [x for x in range(0,len(IO_model_transfer),2)]
    plotly.offline.iplot({
            "data": [Scatter(x=[x for x in range(len(IO_model_transfer[0:size]))] , y=IO_model_transfer[0:size],name='Output gate'),
                    Scatter(x=[x for x in range(len(IO_model_baseline[0:size]))] , y=IO_model_baseline[0:size],name='Baseline'),
                    Scatter(x=[x for x in range(len(I_model_transfer[0:size]))] , y=I_model_transfer[0:size],name='No output gate')],
                    
                    
            "layout": Layout(title="Error Curves for Red-Green-Purple Gridworld Task",
                    xaxis=dict(title="# of trials",ticktext=ticktext, tickvals=tickvals),
                    yaxis=dict(title="Mean Error (Actual Steps - Optimal Steps)"))
            })
    
    I_model_transfer = list(map(float,I_model_transfer))
    IO_model_transfer = list(map(float,IO_model_transfer))
    IO_model_baseline = list(map(float,IO_model_baseline))
    print("I_model:",min(I_model_transfer))
    print("IO_model:",min(IO_model_transfer))
    print("IO_model:",min(IO_model_baseline))
    threshold = 1
    I_model_t_count = None
    IO_model_t_count = None
    IO_base_t_count = None
    flag1,flag2,flag3 = 1,1,1
    for x in range(len(I_model_transfer)):
        if I_model_transfer[x] <= threshold and flag1:
            I_model_t_count = x
            flag1 = 0
        if IO_model_transfer[x] <= threshold and flag2:
            IO_model_t_count = x
            flag2 = 0
        if IO_model_baseline[x] <= threshold and flag3:
            IO_base_t_count = x
            flag3 = 0
    print('I_model:',I_model_t_count)
    print('IO_model:',IO_model_t_count)
    print('IO_base_t_count:',IO_base_t_count)
    
    
    # get jumpstart
    x1 = I_model_transfer[0]
    x2 = IO_model_transfer[0]
    x3 = IO_model_baseline[0]
    print("I_Gate:",x1)
    print("IO_Gate:",x2)
    print("Baseline:",x3)
    print('Difference:',abs(x1-x2))
    print('Base_Diff:',abs(x2-x3))
    trace1 = Bar(
    x=['Output gating','No output gating','Baseline'],
    y=[x2, x1, x3],
    name='Output gating',
    marker=dict(
        color=['red','lightblue','lightgreen'],
        line=dict(
            color=[],
            width=2
        )
    )
    )
    data = [trace1]
    layout = Layout(
        title = 'Initial Error (Jumpstart) for Red-Green-Purple Gridworld Task',
        barmode='group',
        xaxis = dict(
            title='Models'
        ),
        yaxis = dict(
            title='Mean Error (Optimal steps - Actual steps)'
        )
    )

    fig = Figure(data=data, layout=layout)
    plotly.offline.iplot(fig, filename='grouped-bar')
```

```python
create_transfer_graph()
```
